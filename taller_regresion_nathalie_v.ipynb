```python
import pandas as pd
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
groups = ["Texas", "Boston", "Detroit", "Kansas", "St.", "New_S.", "NeW_Y.",
           "Milwaukee", "Colorado", "Houston", "Baltimore", "Los_An.",
           "Chicago", "Cincinnati", "Los_P.", "Philadelphia", "Chicago",
           "Cleveland", "Arizona", "Toronto", "Minnesota", "Florida",
           "Pittsburgh", "Oakland", "Tampa", "Atlanta", "Washington",
           "San.F", "San.I", "Seattle"]

numbers = [5659, 5710, 5563, 5672, 5532, 5600, 5518, 5447, 5544, 5598,
                 5585, 5436, 5549, 5612, 5513, 5579, 5502, 5509, 5421, 5559,
                 5487, 5508, 5421, 5452, 5436, 5528, 5441, 5486, 5417, 5421]

runs = [855, 875, 787, 730, 762, 718, 867, 721, 735, 615, 708, 644, 654, 735,
        667, 713, 654, 704, 731, 743, 619, 625, 610, 645, 707, 641, 624, 570,
        593, 556]
```

```python
data = pd.DataFrame({'groups': groups, 'numbers': numbers, 'runs': runs})

X_stats_models = sm.add_constant(data['numbers'])
model_stats_models = sm.OLS(data['runs'], X_stats_models).fit()
```

```python
print("Result Stats Models:")
print(model_stats_models.summary())
print("\n")
```

```python
X_scikit_learn = data[['numbers']]
y_scikit_learn = data['runs']
model_scikit_learn = LinearRegression()
model_scikit_learn.fit(X_scikit_learn, y_scikit_learn)
```

```python
print("Result de scikit-learn:")
print('Coeficiente:', model_scikit_learn.coef_[0])
print('Intercepto:', model_scikit_learn.intercept_)
print("\n")
```

```python
predicciones_scikit_learn = model_scikit_learn.predict(X_scikit_learn)
```

```python
plt.scatter(X_scikit_learn, y_scikit_learn, color='blue', label='Real Data')
plt.plot(X_scikit_learn, predicciones_scikit_learn, color='red', linewidth=2, label='Lineal Regression')
plt.xlabel('Number bate')
plt.ylabel('Runs')
plt.title('Lineal Regression: Number bate vs. Carreras')
plt.legend()
plt.show()
```
